```python
# DFS

tree = {
    "id": 1,
    "val": 1,
    "left": {
        "id": 2,
        "val": 2,
        "left": {
            "id": 4,
            "val": 4
        },
        "right": {
            "id": 5,
            "val": 5
        }
    },
    "right": {
        "id": 3,
        "val": 3
    }
}

# I build a recursive call in depth, and depeding on order of value, left and right I can get different
# combinations of ordering. It is recursive, so it will go in depth
def dfs(node):
    if node is None:
        return
    print(node['val'])
    dfs(node.get('left'))
    dfs(node.get("right"))


print(dfs(tree))
```

```python
# BFS

tree = {
    "id": 1,
    "val": 1,
    "left": {
        "id": 2,
        "val": 2,
        "left": {
            "id": 4,
            "val": 4
        },
        "right": {
            "id": 5,
            "val": 5
        }
    },
    "right": {
        "id": 3,
        "val": 3
    }
}

# I'm adding a new LIFO stack at the beginning with firts element
# then I pop the first, print the value and find left and rigth and add them to stack
# I repeat until no elements left

def bfs(root):
    if root is None:
        return

    queue = [root] 

    while queue:
        node = queue.pop(0) 
        print(node["val"])

        if node.get("left"):
            queue.append(node["left"])
        if node.get("right"):
            queue.append(node["right"])


print(bfs(tree))
```
