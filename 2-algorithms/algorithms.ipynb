```python
# Binary Search

# Complexity: O(log(N))

# Task: [1, 2, 3, 4, 5, 6] -> [0 ... 5] -> find 'search'
# Base Case:
# - if left > right -> return False
# Recursive case:
# - pick middle (arithmetic mean), compare
# - if middle is bigger than 'search' -> move left pointer to middle + 1 (we need to exclude middle)
# - if middle is smaller than 'search' -> move right pointer to middle - 1 (we need to exclude middle)

arr = [1, 2, 3, 4, 5, 6]


def binary_search(search, arr, left, right):
    if left > right:
        return False

    middle = (left + right) // 2

    if (arr[middle] == search):
        return True

    if (arr[middle] > search):
        return binary_search(search, arr, left, middle - 1)
    else:
        return binary_search(search, arr, middle + 1, right)


binary_search(1, arr, 0, len(arr) - 1)
```
