```python
# Binary Search

# Complexity: O(log(N))

# Task: [1, 2, 3, 4, 5, 6] -> [0 ... 5] -> find 'search'
# Base Case:
# - if left > right -> return False
# Recursive case:
# - pick middle (arithmetic mean), compare
# - if middle is bigger than 'search' -> move left pointer to middle + 1 (we need to exclude middle)
# - if middle is smaller than 'search' -> move right pointer to middle - 1 (we need to exclude middle)

arr = [1, 2, 3, 4, 5, 6]


def binary_search(search, arr, left, right):
    if left > right:
        return False

    middle = (left + right) // 2

    if (arr[middle] == search):
        return True

    if (arr[middle] > search):
        return binary_search(search, arr, left, middle - 1)
    else:
        return binary_search(search, arr, middle + 1, right)


binary_search(1, arr, 0, len(arr) - 1)
```

```python
# Quick sort

# Complexity: O(N log N) to O(N**2)

# Task: [2, 1, 100, -1000] -> [-1000, 1, 2, 100]
# Base Case:
# - if array is 0 or 1 length -> return
# Recursive case:
# - pick first, create array 'before' and array 'after'
# - return concatenated array with recursive chain

arr = [2, 1, 100, -1000]


def quick_sort(arr):

    if (len(arr) <= 1):
        return arr

    pivot = arr[0]
    left = []
    right = []

    for el in arr[1:]:
        if (el <= pivot):
            left.append(el)

        else:
            right.append(el)

    return quick_sort(left) + [pivot] + quick_sort(right)


result = quick_sort(arr)
print(result)
```

```python
# Factorial recursive and tabulating

# Complexity: O(N)

# Recursive is from top to bottom
# Tabulation is from bottom to top (we are building a table with size 1)

data = 5


def factorial_recursive(data):
    if data == 0:
        return 1
    else:
        return data * factorial_recursive(data - 1)


def factorial_tabulated(data):
    result = 1
    increment = 2

    while (increment <= data):
        result *= increment
        increment += 1

    return result


result_1 = factorial_recursive(data)
result_2 = factorial_tabulated(data)

print(result_1)
print(result_2)
```
