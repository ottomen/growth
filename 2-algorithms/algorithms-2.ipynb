```python
# Prefix sum

# ---------------------------------------------------------------
# Calculate Bus Occupancy at Each Stop
# ---------------------------------------------------------------
# You are given:
# 1. A total number of bus stops (stops), numbered sequentially from 1 to stops.
# 2. A list of tickets, where each ticket is a tuple [start, stop] indicating:
#    - start: the stop where the passenger boards the bus
#    - stop: the stop where the passenger leaves the bus
# Your goal is to get the number of passengers on the bus at each stop throughout the route.

stops = 6
tickets = [[1, 4], [2, 5]]


def get_occupancy(stops, tickets):
    hash = {}
    prefixSum = 0
    result = [0 for _ in range(stops)]

    for start, stop in tickets:
        hash[start - 1] = hash.get(start - 1, 0) + 1
        hash[stop - 1] = hash.get(stop - 1, 0) - 1

    for i in range(stops):
        if i in hash:
            prefixSum += hash[i]

        result[i] += prefixSum

    return result


# [1,2,2,1,0,0]
print(get_occupancy(stops, tickets))
```

```python
# DFS

tree = {
    "id": 1,
    "val": 1,
    "left": {
        "id": 2,
        "val": 2,
        "left": {
            "id": 4,
            "val": 4
        },
        "right": {
            "id": 5,
            "val": 5
        }
    },
    "right": {
        "id": 3,
        "val": 3
    }
}

# I build a recursive call in depth, and depeding on order of value, left and right I can get different
# combinations of ordering. It is recursive, so it will go in depth
def dfs(node):
    if node is None:
        return
    dfs(node.get('left'))
    print(node['val'])
    dfs(node.get("right"))


print(dfs(tree))
```

```python
# BFS

tree = {
    "id": 1,
    "val": 1,
    "left": {
        "id": 2,
        "val": 2,
        "left": {
            "id": 4,
            "val": 4
        },
        "right": {
            "id": 5,
            "val": 5
        }
    },
    "right": {
        "id": 3,
        "val": 3
    }
}

# I'm adding a new LIFO stack at the beginning with firts element
# then I pop the first, print the value and find left and rigth and add them to stack
# I repeat until no elements left

def bfs(root):
    if root is None:
        return

    queue = [root] 

    while queue:
        node = queue.pop(0) 
        print(node["val"])

        if node.get("left"):
            queue.append(node["left"])
        if node.get("right"):
            queue.append(node["right"])


print(bfs(tree))
```
